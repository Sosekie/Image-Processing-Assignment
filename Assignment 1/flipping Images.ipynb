```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open('Cervin.png')
```

```python
image_array = np.array(image)
height, width, channels = image_array.shape
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(141)
plt.imshow(image)
plt.title('Original Image')
```

```python
horizontal_flipped_image_array = np.zeros_like(image_array)
for i in range(height):
    for j in range(width):
        # Putting the elements from the right side of the matrix to the left side
        horizontal_flipped_image_array[i, j] = image_array[i, width - j - 1]

horizontal_flipped_image = Image.fromarray(horizontal_flipped_image_array)
horizontal_flipped_image.save('horizontal_flip.png')

plt.subplot(142)
plt.imshow(horizontal_flipped_image)
plt.title('Horizontal Flip')
```

```python
vertical_flipped_image_array = np.zeros_like(image_array)
for i in range(height):
    for j in range(width):
        # Putting the elements from the bottom of the matrix to the top
        vertical_flipped_image_array[i, j] = image_array[height - i - 1, j]

vertical_flipped_image = Image.fromarray(vertical_flipped_image_array)
vertical_flipped_image.save('vertical_flip.png')

plt.subplot(143)
plt.imshow(vertical_flipped_image)
plt.title('Vertical Flip')
```

```python
both_flipped_image_array = np.zeros_like(image_array)
for i in range(height):
    for j in range(width):
        # Combine of putting
        both_flipped_image_array[i, j] = image_array[height - i - 1, width - j - 1]

both_flipped_image = Image.fromarray(both_flipped_image_array)
both_flipped_image.save('both_flip.png')

plt.subplot(144)
plt.imshow(both_flipped_image)
plt.title('Both Flip')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

axes[0, 0].imshow(image_array)
axes[0, 0].set_title('Original image')

axes[0, 1].imshow(horizontal_flipped_image)
axes[0, 1].set_title('Horizontal Flip')

axes[1, 0].imshow(vertical_flipped_image)
axes[1, 0].set_title('Vertical Flip')

axes[1, 1].imshow(both_flipped_image)
axes[1, 1].set_title('Both Flip')
```
